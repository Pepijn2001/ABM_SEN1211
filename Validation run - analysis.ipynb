## Timer logic

```python
# Look at a random agent
agent_data[agent_data['AgentID']==79]
```

As can be seen in the dataframe above, the timers go down by one each time step, just like they are supposed to. As per the last else part of the normal_behaviour function; the countdown of timer A should freeze the moment timer B goes off vise versa. In the data above, it becomes clear that this works around timestep 167. Looking at the next step - the moment the toilet timer of the agent goes off, it stops looking at its friends and starts moving towards the toilet. The target location is set on the locations of the toilets in the map, and it is visible that each time step, the agents moves closer towards its goal. It is also visible that the agents saves its previous location/ current location each time step. The moment the timer goes off, the previous location does not get updated anymore. This way the agent can return to the location it was whenever it is done doing its tasks. At timestep 208 the agent reaches the toilets. This can be seen by the current location that matches the target location, as well as the shift in state (from moving towards the toilet to waiting at toilet). From that point on the toilet delay also starts counting down, indicating that the code works. As the validation run was only done with 300 time steps, this won't be shown anymore.

```python
# Look at another random agent
agent_data[agent_data['AgentID']==54]
```

That the code for returning to the previous location works can be seen when looking at agent 54. At step 7 its bar timer goes of, and since it is already standing close to the bar, it reaches the bar at step 14. The agent has a long bar delay, at time step 150 this delay is done. As can be seen by the change in state as well as the target location, it shows that the agents moves back to where it came from. The moment the agent reaches its old location (where it was before the bar timer went off), the target location resets and it starts looking for friends.

## Find friends function
Agent 54 can also be used to show the workings of the find friends function. The moment the agent returns to the location it was when its bar timer went off (Setp 154), it starts looking for its friends. This agent has only one friend - agent 55. Looking at the columns for friends in vision far and close, it becomes clear that the friend is visible in the far away vision, but not in the close one. The columns for Agents in vision far and close serve as an additional check to see if the code works - agent 55 is part of all of the agents agent 54 can see in its far away vision, but not in the close by vision. This means that agent 54 has to move closer towards agent 55 as to not loose them. Looking at the state variable, this also happens; agent 55 notices that its friend far away, so it is moving closer. Twice the state 'Friends far away - not visible' is shown. This can be the case because agent 54 (the friend) is also moving. If it happens to move behind an obstacle, agent 55 cannot see it, and thus cannot move towards it (it will move randomly). Luckily, the friend are reunited shortly afterwards in step 159. The agents are close enough together that they are moving randomly (or dancing as you will) instead of towards each other. This distance between the friends is evaluated each time step to make sure that they dont move to far away from each other and get left alone. 

## Stair target

```python
# Random agent that takes the stairs
agent_data[agent_data['AgentID']==11]
```

The agent above shows the workings of the take stars function. Around step 40, the agents toilet timer goes off. The moment this happens, the code checks at what floor the agent currently is, and on what floor their target location is. In this case, the location of the toilet is on the ground floor, while the agent is at the first floor. The agent thus can't directly move towards its target (as it will get stuck walking against a wall), but has to take the stair first. On step 41, this check is done, and an additional target location (Stair target) is added to the agent. The agent first walks towards the stairs - as can be seen by following its current location. When it has reached the stair target, it is transported to the fround floor (step 47), the stair target disappears, and the agent can continue its way towards the toilet.

## Search timer & observer

```python
agent_data_evac[agent_data_evac['AgentID']==57]
```

Social agents will first try and find their friends, before evacuating. However, if after a certain time (search timer) they havnt found any of their friends, they will move towards an exit. Depending on if the social agents is an observer or not, they will either look for the closest exit, or move towards a known exit (which is the main entrance). Agent 57 above is a social observer. As can be seen in the dataframe, the agent moves randomly trying to find their friend. Around step 20, when the search timer has gone off, the agent still hasnt found their friend. As this agent is an observer, it starts looking around for an exit while moving randomly. They find one at once, and move towards it. However, when later it spots a crew member, it adapts their target location to the closer exit (which they have been made aware of by the crew, for more see below).

## Seen crew

```python
agent_data_evac[agent_data_evac['AgentID']==7]
```

Agent 7 above shows the importance of the crew even more. This agent is with their friend the moment the alarm goes of. Both of its friends are socials but not observers (see below, only agent 5 is visible for agent 7, but the agent reporter was not implemented correctly so the column does not give a value), meaning that the moment they have spotted a friend, they will walk towards a known exit together, which is the main entrance. This is also visible in the current location column; the moment the agent has found a friend, it takes over their target location and moves towards it.  (As is inherent with ABM models, there is an order of education. In this case, agent 5 does all of their steps before agent 7. Agent 5 has found a friend, and as they are not an observer, sets their target location to the main entrance. Agent 7 does their steps afterwards - they have found their friend agent 5, which has a target location. Agent 7 takes over this target location.). Immediately after, they run into a crew member, which points them towards a closer exit (see the change in state and target location), ensuring that the agents can evacuate as fast and safe as possible.
